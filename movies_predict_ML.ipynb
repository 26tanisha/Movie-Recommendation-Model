```python
import ast
def fetch_director(obj):
    try:
        crew_list = ast.literal_eval(obj)
        for i in crew_list:
            if i.get('job') == 'Director':
                return i.get('name')
        return None
    except:
        return None
```

```python
import ast

def clean_crew(x):
    # Case 1: if it's already a list
    if isinstance(x, list):
        # If it's a list of characters like ['J', 'a', 'm'...] — ignore it
        if all(isinstance(i, str) and len(i) == 1 for i in x):
            return []
        return [i.replace(" ", "") for i in x]

    # Case 2: if it's a string
    if isinstance(x, str):
        try:
            # Try to evaluate it as a list
            evaluated = ast.literal_eval(x)
            if isinstance(evaluated, list):
                return [i.replace(" ", "") for i in evaluated]
        except:
            pass
        # Plain string case like 'James Cameron'
        return [x.replace(" ", "")]

    # Fallback for anything else
    return []

movies['crew'] = movies['crew'].apply(clean_crew)
# "[ 'James Cameron', 'Jon Landau' ]"   → string
# "James Cameron"                       → string
# ['James Cameron', 'Jon Landau']       → list
# ['J','a','m','e','s']                 → wrong list
# convert all into list 
```

```python
def combine_tags(row):
    return (
        str(row['id']) + " " +
        str(row['title']) + " " +
        " ".join(row['genres']) + " " +
        str(row['overview']) + " " +
        " ".join(row['keywords']) + " " +
        " ".join(row['cast']) + " " +
        " ".join(row['crew'])
    )

movies['tags'] = movies.apply(combine_tags, axis=1)

```

```python
def recommend(movie):
    if movie not in new_df['title'].values:
        print(f"'{movie}' not found in database.")
        return

    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)

```

```python
recommend('Batman Begins')
```
